Feature Understanding (Univariate analysis)

```python
ax = df['Year_Introduced'].value_counts().head(10).plot(kind='bar',title='Top Years Coasters Introduced')
ax.set_xlabel('Year introduced')
ax.set_ylabel('Count')
```

```python
ax = df['Speed_mph'].plot(kind='hist',bins=20,title='Coaster speed (mph)')
ax.set_xlabel('Speed(mph)')
```

```python
ax = df['Speed_mph'].plot(kind='kde',title='Coaster speed (mph)')
ax.set_xlabel('Speed(mph)')
```

Feature relationships 

```python
df.plot(kind='scatter',x='Speed_mph',y='Height_ft')
plt.show()
```

```python
sns.scatterplot(x='Speed_mph',y='Height_ft',data=df,hue='Year_Introduced')
plt.show()
```

```python
#Comparing more than one feature 
sns.pairplot(df,vars=['Year_Introduced','Speed_mph','Height_ft','Inversions','Gforce'],hue='Type_Main')
plt.show()
```

```python
df_corr = df[['Year_Introduced','Speed_mph','Height_ft','Inversions','Gforce']].dropna().corr()
df_corr 
```

```python
sns.heatmap(df_corr,annot=True)
```

```python
df.query('Location != "Other"').groupby('Location')['Speed_mph'].agg(['mean','count']).query('count >= 10').sort_values('mean')['mean'].plot(kind='barh',figsize=(12,8),title='Average Coaster speed by location')
ax.set_xlabel('Average Coaster speed')
```
